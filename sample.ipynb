## Query the graph created in "Set up and query Spanner Graph"

[Set up and query Spanner Graph] page contains a finance graph used throughout the documentation. Replace the placeholder values of `{project_name}`, `{instance_name}` and `{database_name}` in the query below to visualize nodes and edges from that graph.

```python
%%spanner_graph --project {project_name} --instance {instance_name} --database {database_name}

GRAPH FinGraph
MATCH (a)-[e]->(b)
RETURN TO_JSON(a) AS src_node,
       TO_JSON(e) AS edge,
       TO_JSON(b) AS dst_node
```
